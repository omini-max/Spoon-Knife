<h2>Data Wrangling</h2>


<h3>Import data</h3>
<p>
The Data is an "Automobile Data Set" and can be retrieved from the following link: <a href="https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data">https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data</a>.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

<h2> Reading the URL of the Automobile dataset and adding headers</h2>

```python
filename = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

Use the method <b>read_csv()</b> to load the data in to pandas DataFrame, set names to headers

```python
df = pd.read_csv(filename, names= headers)
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

Deal with the missing data

```python
import numpy as np
# replace missing data with NaN
df.replace("?", np.nan, inplace = True)
df.head()
```

use the method <b>isnull()</b>, to identify missing data, <b>value_count()</b> to count missing data in each column

```python
missing_data = df.isnull()
missing_data.head()
```

```python
for column in missing_data:
  print(column)
  print(missing_data[column].value_counts())
  print("")
```

Replace missing data with means, mode, or we drop the row with missing data.

```python
mean_norm_loss = df['normalized-losses'].astype("float").mean(axis=0)
print("Average normalized-losses: ",mean_norm_loss)
df['normalized-losses'].replace(np.nan, mean_norm_loss, inplace = True)

mean_bore = df['bore'].astype("float").mean(axis=0)
print("Average bore: ",mean_bore)
df['bore'].replace(np.nan, mean_bore, inplace = True)

mean_stroke = df['stroke'].astype("float").mean(axis=0)
print("Average stroke: ",mean_stroke)
df['stroke'].replace(np.nan, mean_stroke, inplace = True)

mean_horsepower = df['horsepower'].astype("float").mean(axis=0)
print("Average horsepower: ",mean_horsepower)
df['horsepower'].replace(np.nan, mean_horsepower, inplace = True)

mean_peak_rpm = df['peak-rpm'].astype("float").mean(axis=0)
print("Average peak-rpm: ",mean_peak_rpm)
df['peak-rpm'].replace(np.nan, mean_peak_rpm, inplace = True)
df.head()
```

since the <b>num-of-doors<b> column is of object type, the mode is used, as <b>price<b> is the target and what we want to predict, we drop rows with missing values.

```python
print(df['num-of-doors'].value_counts())
df['num-of-doors'].replace(np.nan, "four", inplace =True)
df.dropna(subset=['price'], axis = 0, inplace = True )
df.reset_index(drop = True, inplace = True)
df.head()
```

To confirm each column has the right data type <b>.dtypes</b> and <b>.astype</b> to set the correct data type.

```python
df.dtypes
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
df[["horsepower"]] = df[["horsepower"]].astype("int")
```

```python
df.dtypes
```

Data in the city-mpg and highway-mpg can be converted to L/km depending on standards.
the columns can be renamed accordingly, after conversion.

feature scaling to normalize the data with wide ranges.

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
df[['length', 'width', 'height']].head()
```

```python
df.head(10)
```

<h3>Standardization</h3>
<p>Data in the mpg columns should be changed based on the standard in the region the analysis is needed.
hence, mpg will be changed to L/100km.
L/100km = 235/mpg.</p>


```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df["city-mpg"] = 235/df["city-mpg"]
df.rename(columns={'city-mpg':'city-L/100km'}, inplace=True)
df["highway-mpg"]=235/df["highway-mpg"]
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)
# check your transformed data
df.head()
```

```python
df.to_csv('clean_auto_df.csv')
```
